```python
import pandas as pd
import re

# Load dataset
df = pd.read_csv("spam.csv", encoding='latin-1')[['v1', 'v2']]
df.columns = ['label', 'text']

# Clean text
def clean_text(text):
    text = text.lower()                              # 1. lowercase
    text = re.sub(r'[^\w\s]', '', text)              # 2. remove punctuation
    text = re.sub(r'\d+', '', text)                  # 3. remove numbers
    text = re.sub(r'\s+', ' ', text).strip()         # 4. remove extra spaces
    return text

df['text'] = df['text'].apply(clean_text)


```

```python
df.head()
```

# 2. Bag of Words (BoW)

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform(df['text'])

```

```python
print(vectorizer.get_feature_names_out()[:30])
print(X_bow.toarray()[0][:30])  # First row
```

# . TF-IDF (Term Frequency - Inverse Document Frequency)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df['text'])

```

```python
print(tfidf.get_feature_names_out()[:10])
print(X_tfidf.toarray()[0][:10])
```

```python
import pandas as pd
from transformers import BertTokenizer, BertModel
import torch
from tqdm import tqdm

# Step 1: Load dataset (first 100 rows)
df = pd.read_csv("spam.csv", encoding='latin-1')[['v1', 'v2']]
df.columns = ['label', 'text']
df = df.dropna().head(100)

# Step 2: Load BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Step 3: Function to get CLS vector
def get_bert_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
    cls_embedding = outputs.last_hidden_state[:, 0, :]  # CLS token (sentence embedding)
    return cls_embedding.squeeze().numpy()

# Step 4: Apply to all 100 texts
tqdm.pandas()
df['bert_vector'] = df['text'].progress_apply(get_bert_embedding)

# Step 5: Check one result
print(df[['text', 'bert_vector']].head())
```
